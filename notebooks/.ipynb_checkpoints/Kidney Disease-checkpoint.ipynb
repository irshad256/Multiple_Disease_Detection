```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import confusion_matrix,roc_auc_score
from sklearn.metrics import classification_report
from sklearn.impute import KNNImputer
import pickle
```

```python
df=pd.read_csv('../dataset/kidney_disease.csv')
df.head()
```

```python
df.classification=df.classification.replace('ckd\t','ckd')
```

```python
df.pcv.unique()
```

```python
df.rc.unique()
```

```python
df.wc.unique()
```

```python
df.pcv=df.pcv.str.replace(r'\t','')
df.rc=df.rc.str.replace(r'\t','')
df.wc=df.wc.str.replace(r'\t','')
df=df.replace(r'?',np.nan)
df.pcv=df.pcv.astype('float')
df.rc=df.rc.astype('float')
df.wc=df.wc.astype('float')
```

```python
sns.countplot(x='classification',data = df)
```

```python
df.drop(['id'],axis=1,inplace=True)
```

```python
lis = []
for i in range(0,df.shape[1]):
     
    if(df.iloc[:,i].dtypes == 'object'):
        df.iloc[:,i] = pd.Categorical(df.iloc[:,i])
        df.iloc[:,i] = df.iloc[:,i].cat.codes 
        df.iloc[:,i] = df.iloc[:,i].astype('object')
         
        lis.append(df.columns[i])
```

```python
KNN=KNNImputer()
```

```python
df = pd.DataFrame(KNNImputer().fit_transform(df), columns = df.columns)
```

```python
df.head()
```

```python
plt.figure(figsize = (20,20))
sns.heatmap(df.corr(), annot = True)
```

```python
X = df.drop(['classification'],axis=1)
y = df['classification']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 10)
```

```python
print("Train Set: ", X_train.shape, y_train.shape)
print("Test Set: ", X_test.shape, y_test.shape)
```

```python
n_estimators=[10,20,30,50,100,200,300,400,500]
for n in n_estimators:
    model=RandomForestClassifier(n_jobs=-1,n_estimators=n,random_state=10)
    model.fit(X_train,y_train)
    y_pred=model.predict(X_test)
    print('n_estimator : ',n,' AUC Score : ',roc_auc_score(y_test,y_pred))
```

```python
classifier=RandomForestClassifier(n_estimators=20,random_state=10)
classifier.fit(X_train, y_train)
confusion_matrix(y_test, classifier.predict(X_test))
```

```python
y_pred=classifier.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
print(f"ROC_AUC score is {round(roc_auc_score(y_test, classifier.predict(X_test))*100,2)}%")
```
